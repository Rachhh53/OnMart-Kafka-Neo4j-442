<div class="alert alert-warning">
<br>
The idea behind using the LSTM model is to input sequences of products bought and to predict products the user might be interested in buying next. The below picture summarizes the methodology, insted of predicting the next word in a sequence, here we predict the next product bought. 
</br> 

![alt text](../images/nlp_vs_seq.png "Title")

### Inputs to the model:
- Sequence containing the entire history of products bought for a particular user
- Input Shape: [18] (max len of a sequence)

### Hidden layers:
- **Embedding layer:** We use an embedding layer to fix a constant size of inputs for learning.
- **Stacked LSTM layer:** Stacked LSTMs are now a stable technique for challenging sequence prediction problems. A Stacked LSTM architecture can be defined as an LSTM model comprised of multiple LSTM layers. An LSTM layer above provides a sequence output rather than a single value output to the LSTM layer below. Specifically, one output per input time step, rather than one output time step for all input time steps. We use **64** stacked LSTM layers for our model.
- **Self attention layer**: A self-attention module takes in n inputs, and returns n outputs. What happens in this module? In layman’s terms, the self-attention mechanism allows the inputs to interact with each other (“self”) and find out who they should pay more attention to (“attention”). The outputs are aggregates of these interactions and attention scores.

### Output of the model:
- Recommendation scores for the input sequence for each product  
- Output shape: [18 x 6] (predictions for all the 18 places in the sequence, 5 products + 1 padding) 

### Loss for the model:
- Modified Sparse categorical cross entropy. (Loss funtion for a classification model)
</div>

```python
# Vectorization
transaction_tf_df = pd.DataFrame()
transaction_tf_df['CustomerID'] = transactions_wo_high_returns['CustomerID'].astype(str)
transaction_tf_df['Product_Name'] = transactions_wo_high_returns['Product_Name'].astype(str)
transactionsdf = transactionsdf.dropna()
ord_enc = OrdinalEncoder()
transaction_tf_df["Product_Name_encoded"] = ord_enc.fit_transform(transaction_tf_df[["Product_Name"]]).astype(int)
transaction_tf_df["Product_Name_encoded"] = transaction_tf_df["Product_Name_encoded"].apply(lambda x:(x+1))
transaction_tf_df.head()
```

```python
def list_(ls):
    """
    Create a function to return a sequence separated by commas
    :return: String containing the sequence of products bought by each customer
    """
    return ",".join(map(str, ls))
```

```python
# Create Sequences:
merged_transaction_tf_df = transaction_tf_df.groupby("CustomerID").agg({"Product_Name_encoded": ['count',list_]})
merged_transaction_tf_df = merged_transaction_tf_df.transpose().reset_index(level=0, drop=True).transpose().reset_index()
merged_transaction_tf_df.columns = ['CustomerID','Products_count','Product_ordered']
# Prepare the dataframe for training:
merged_transaction_tf_df = merged_transaction_tf_df.drop(columns=['Products_count'])
# Convert Products_orderd sequence to list:
merged_transaction_tf_df['Product_ordered'] = merged_transaction_tf_df['Product_ordered'].apply(lambda x: list(map(float, x.split(','))))
# Remove the last index of products bought and place it in the target column
merged_transaction_tf_df['Products_for_training'] = merged_transaction_tf_df['Product_ordered'].apply(lambda x: x[:len(x)-1])
merged_transaction_tf_df['Target'] = merged_transaction_tf_df['Product_ordered']
# Drop columns that are not needed for training
merged_transaction_tf_df.head()
```

```python
merged_transaction_tf_df
```

```python
dict_transaction_tf_df = merged_transaction_tf_df.to_dict('list')
train_dict = {'Products_for_training': dict_transaction_tf_df['Products_for_training'],
              'Target': dict_transaction_tf_df['Target']} 
train_dict.items()
```

```python
# Padding function
def padding(seq):
#   Max length = Maximum products bought by one user
    return tf.keras.preprocessing.sequence.pad_sequences(
            train_dict[seq], maxlen=18, dtype='float32', padding='pre',
            truncating='pre', value=0.0)

# form a dictionary for training
labels = ['Products_for_training','Target']
padded_sequences = []
for item in labels:
    padded_sequences.append(padding(item))
    
# Map respective sequences to its labels
padded_train_dict = {'Products_for_training': padded_sequences[0],
                     'Target': padded_sequences[1]} 
padded_train_dict.items()
```

```python
def create_train_tfdata(train_feat_dict, train_target_tensor,
                        batch_size, buffer_size=None):
    """
    Create train tf dataset for model train input
    :param train_feat_dict: dict, containing the features tensors for train data
    :param train_target_tensor: np.array(), the training TARGET tensor
    :param batch_size: (int) size of the batch to work with
    :param buffer_size: (int) Optional. Default is None. Size of the buffer
    :return: (tuple) 1st element is the training dataset,
                     2nd is the number of steps per epoch (based on batch size)
    """
    if buffer_size is None:
        buffer_size = batch_size*50

    train_steps_per_epoch = len(train_target_tensor) // batch_size

    train_dataset = tf.data.Dataset.from_tensor_slices((train_feat_dict,
                                                        train_target_tensor)).cache()
    train_dataset = train_dataset.shuffle(buffer_size).batch(batch_size)
    train_dataset = train_dataset.repeat().prefetch(tf.data.experimental.AUTOTUNE)
    
    return train_dataset, train_steps_per_epoch
  
train_feat_dict = {'Products_for_training': padded_train_dict['Products_for_training']}
train_target_tensor = padded_train_dict['Target']
train_dataset, train_steps_per_epoch = create_train_tfdata(train_feat_dict,
                                                             train_target_tensor,
                                                             batch_size=512)  
```

```python
def loss_function(real, pred):
    """
    We redefine our own loss function in order to get rid of the '0' value
    which is the one used for padding. This to avoid that the model optimize itself
    by predicting this value because it is the padding one.
    
    :param real: the truth
    :param pred: predictions
    :return: a masked loss where '0' in real (due to padding)
                are not taken into account for the evaluation
    """

    # to check that pred is numric and not nan
    mask = tf.math.logical_not(tf.math.equal(real, 0))
    loss_object_ = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True,
                                                                 reduction='none')
    loss_ = loss_object_(real, pred)
    mask = tf.cast(mask, dtype=loss_.dtype)
    loss_ *= mask

    return tf.reduce_mean(loss_)
```

```python
def build_model():
    """
    Build a model given the hyper-parameters with item and nb_days input features
    :return: built and compiled tensorflow model 
    """
    
    Product_sequences = tf.keras.Input(batch_input_shape=[None,  18],
                                       name='Products_for_training', dtype=tf.float32)
    # create encoding padding mask
    encoding_padding_mask = tf.math.logical_not(tf.math.equal(Product_sequences, 0))

    # Pass categorical input through embedding layer
    embedding_item = tf.keras.layers.Embedding(input_dim=6, # updated to 6 as we only have 5 products
                                               output_dim=64,
                                               name='embedding_item'
                                              )(Product_sequences)


    Normalize_embeddings = tf.keras.layers.BatchNormalization(
     name='batchnorm_inputs')(embedding_item)
    
    # LSTM layer
    rnn = tf.keras.layers.LSTM(units=64,
                                   return_sequences=True,
                                   stateful=False,
                                   recurrent_initializer='glorot_normal',
                                   name='LSTM_cat'
                                   )(Normalize_embeddings)

    rnn = tf.keras.layers.BatchNormalization(name='batchnorm_lstm')(rnn)

    # Self attention so key=value in inputs
    att = tf.keras.layers.Attention(use_scale=False, causal=True,
                                    name='attention')(inputs=[rnn, rnn],
                                                      mask=[encoding_padding_mask,
                                                            encoding_padding_mask])

    # Last layer is a fully connected one
    output = tf.keras.layers.Dense(6, name='output')(att)

    model = tf.keras.Model(Product_sequences, output)

    model.compile(
        optimizer=tf.keras.optimizers.Adam(lr = 0.1),
        loss=loss_function,
        metrics=['sparse_categorical_accuracy'])
    
    return model
```

```python
def fit_model(model, train_dataset, steps_per_epoch, epochs):
    """
    Fit the Keras model on the training dataset for a number of given epochs
    :param model: tf model to be trained
    :param train_dataset: (tf.data.Dataset object) the training dataset
                          used to fit the model
    :param steps_per_epoch: (int) Total number of steps (batches of samples) before 
                            declaring one epoch finished and starting the next epoch.
    :param epochs: (int) the number of epochs for the fitting phase
    :return: tuple (mirrored_model, history) with trained model and model history
    """
    
    # mirrored_strategy allows to use multi GPUs when available
    mirrored_strategy = tf.distribute.experimental.MultiWorkerMirroredStrategy(
        tf.distribute.experimental.CollectiveCommunication.AUTO)
    
    with mirrored_strategy.scope():
        mirrored_model = model

    history = mirrored_model.fit(train_dataset,
                                 steps_per_epoch=steps_per_epoch,
                                 epochs=epochs, verbose=2)

    return mirrored_model, history
```
